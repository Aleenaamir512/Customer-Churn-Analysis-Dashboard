```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
df= pd.read_csv("customer_churn_dataset-testing-master.csv")
```

```python
df.columns
```

```python
total_customers= df['CustomerID'].sum()
print(total_customers)

churn_customers=df['Churn'].sum()
print(churn_customers)

active_customers= (total_customers)-(churn_customers)
print(active_customers)

avg_tenure= df['Tenure'].mean()
print(avg_tenure)

churn_rate= df['Churn'].sum()/ (total_customers)
print(churn_rate)
```

```python
print("============KPI=============")
print(f"Total Customers: {total_customers:,.0f}")
print(f"Churn Customers: {churn_customers:,.0f}")
print(f"Active customers: {active_customers:,.0f}")
print(f"Avergae Tenure: {avg_tenure:,.0f}")
print(f"Churn Rate: {churn_rate}")
```

```python
df['Active Users']= (active_customers)
df.columns
```

```python
churn_gender= df.groupby('Gender')['Churn'].sum()
plt.figure(figsize=(6,4))
churn_gender.plot(kind="bar",color="lightgreen")
plt.title("Churn Analysis by Gender")
plt.xlabel("Gender")
plt.ylabel("Churn")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#actibe osers
churn_gender= df.groupby('Gender')['Active Users'].sum()
plt.figure(figsize=(6,4))
churn_gender.plot(kind="bar",color="lightpink")
plt.title("Activity Analysis by Gender")
plt.xlabel("Gender")
plt.ylabel("Active Users")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#churn by subscription
churn_gender= df.groupby('Subscription Type')['Churn'].sum()
plt.figure(figsize=(6,4))
churn_gender.plot(kind="bar",color="brown")
plt.title("Churn by Subscription")
plt.xlabel("Subscription")
plt.ylabel("Churn")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#churn by contract length
contract_length_by_churn= df.groupby('Contract Length')['Churn'].sum()
plt.figure(figsize=(6,4))
contract_length_by_churn.plot(kind="bar",color="lightblue")
plt.title("Contract Length before Churn")
plt.xlabel("Contract Length")
plt.ylabel("Churn")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Load your dataset if not already loaded
# df = pd.read_csv("your_dataset.csv")

# Set plot style (optional)
sns.set(style="whitegrid")

# Create the boxplot
plt.figure(figsize=(6, 5))
sns.boxplot(x='Churn', y='Age', data=df, palette='Set2')

# Add title and labels
plt.title("Age Distribution by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Age")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Optional: load data
# df = pd.read_csv("your_dataset.csv")

# Plot Histogram
plt.figure(figsize=(9, 5))
plt.hist(df['Total Spend'], bins=20, color='skyblue', edgecolor='black')

# Labels and title
plt.title("Distribution of Total Spend")
plt.xlabel("Total Spend Amount")
plt.ylabel("Number of Customers")

# Add grid for clarity
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Replace 1 with 'Yes' and 0 with 'No'
df['Churn'] = df['Churn'].replace({1: 'Yes', 0: 'No'})

# 1. Count of churned vs not churned
churn_counts = df['Churn'].value_counts()

# 2. Labels and values
labels = churn_counts.index
values = churn_counts.values

# 3. Pie chart
plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, 
        wedgeprops={'edgecolor': 'white'})

plt.title("Customer Churn Distribution")
plt.tight_layout()
plt.show()
```
